```python
class GridEnv:
    def __init__(self, num_r, num_c):
        self.num_r = num_r
        self.num_c = num_c
        self.grid = self._build_grid()
        self.start = (0,0)
        self.goal = (4,4)
        self.index = self._index()
        # self.agent = self._create_agent()
        self.i = 0
        self.j = 0

    def __repr__(self):
        return f"grid {self.grid}"

    def _build_grid(self):
        ro = [0] * self.num_c
        grid = [list(ro) for _ in range(self.num_r)]
        walls_def = [[1,1],[1,2],[1,3],[3,1],[3,2],[3,3]]
        
        for i, j in walls_def:
            for row in range(self.num_r):
                for col in range(self.num_c):
                    if (row,col) == (i,j):
                        grid[row][col] = -1
        return grid

    def _index(self):
        index = []
        for i in range(self.num_r):
            for j in range(self.num_c):
                ind = i * self.num_c + j #rot mat
                index.append(ind)

        return index

    def _get_state(self): return self.i * self.num_c + self.j

    def _reset_pos(self):
        self.i = 0
        self.j = 0
        return self._get_state()

    def one_hot(self):
        state = self._get_state()
        tot = self.num_r * self.num_c
        inp = [0] * tot
        inp[state] = 1
        return inp

    def perform(self, act):
        '''
        # self.i = cij[0]
        # self.j = cij[1]
        if act[0] == 'UP':
            if self.i >= 0 and self.j <= self.num_c and self.i <= self.num_r and self.j >= 0:
                self.i -= 1
                self.j = self.j
            else:
                return f"out of bounds"

        if act[0] == 'DOWN':
            if self.i >= 0 and self.j <= self.num_c and self.i <= self.num_r and self.j >= 0:
                self.i += 1
                self.j = self.j
            else:
                return f"out of bounds"
        
        if act[0] == 'RIGHT':
            if self.i >= 0 and self.j <= self.num_c and self.i <= self.num_r and self.j >= 0:
                self.i = self.i
                self.j += 1
            else:
                return f"out of bounds"

        if act[0] == 'LEFT':
            if self.i >= 0 and self.j <= self.num_c and self.i <= self.num_r and self.j >= 0:
                self.i = self.i
                self.j -= 1
            else:
                return f"out of bounds"
        '''
        if act == 1:
            new_i = self.i - 1 #up
            new_j = self.j 

        if act == 2: #down
            new_i = self.i + 1
            new_j = self.j

        if act == 3: #left
            new_i = self.i
            new_j = self.j - 1

        if act == 4: #right
            new_i = self.i
            new_j = self.j + 1

        if new_i < 0 or new_i >= self.num_r or new_j >= self.num_c or new_j < 0:
            reward = -5.0
            state = self._get_state()
            done = False
            return (state, reward, done)

        # if the new pos is val in the grid 
        # ill check if it hit a wall

        if self.grid[new_i][new_j] == -1: #hitting a wall
            reward = -10.0
            done = False
            state = self._get_state()
            return (state, reward, done)

        #if it is not any problem
        #calc its its valid or goal
        #calc trs new pos and reward the NN

        self.i = new_i
        self.j = new_j

        if (new_i, new_j) == self.goal:
            reward = 15.0
            done = True
            state = self._get_state()
            return (state, reward, done)
        else:
            reward = -1.0
            done = False
            state = self._get_state()
            return (state, reward, done)
        
```
